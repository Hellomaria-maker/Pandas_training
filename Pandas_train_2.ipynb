```python
# Анализ данных о бронировании отелей.
import pandas as pd
bookings = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', sep=';')
bookings.shape
bookings.head()
```

```python
bookings_head = bookings[:7]
```

```python
# приведенин названий столбцов к нижнему регистру и замена пробелов на знак _
def replace_space(name):
    new_name = name.replace(' ', '_').lower()
    return new_name

bookings = bookings.rename(columns = replace_space)
bookings.head()
```

```python
# Пользователи из каких стран совершили наибольшее число успешных бронирований? 
# Бронирование считается успешным, если в дальнейшем оно не было отменено (переменная is_canceled). 
bookings.query("is_canceled == 0")\
.country.value_counts()
```

```python
# На сколько ночей (stays_total_nights) в среднем бронируют City Hotel? Resort Hotel?
bookings.groupby("hotel")\
.agg({'stays_total_nights':'mean'}).round(2)
```
